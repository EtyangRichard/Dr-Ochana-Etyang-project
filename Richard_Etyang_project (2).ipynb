#Bar Graph to compare Domestic and foreign Income for Different movie titles.
The bar graph generated reflects data for first 10 samples chosen to represent the wide array of data showing the domestic and foreign income of different movies. This however does not mean that it illustrates the details in descending or ascending order due some details missing. However it is evident from the samples present that a movie generating Large Domestic income is not equivalent to it generating a large foreign income. We therefore aim to pick movies which generate both good foreign and domestic gross income.

````python
# Generating sample data
movies = df['title'][:10]  # Using only the first 10 movies for simplicity
domestic_gross = df['domestic_gross'][:10]
foreign_gross = df['foreign_gross'][:10]

# Number of movies
num_movies = len(movies)

# Bar width
bar_width = 0.35

# Index for each movie
index = np.arange(num_movies)

# Plotting and adjustment of fig size
fig, ax = plt.subplots(figsize=(10, 8))

# Plots for the gross income for domestic and foreign for comparison
bar1 = ax.bar(index - bar_width/2, domestic_gross, bar_width, label='Domestic Gross', color='blue')
bar2 = ax.bar(index + bar_width/2, foreign_gross, bar_width, label='Foreign Gross', color='red')

# Labelling and titling the graph
ax.set_xlabel('Movie Titles')
ax.set_ylabel('Gross Income ($)')
ax.set_title('Domestic and Foreign Gross Income of Movies')
ax.set_xticks(index)
ax.set_xticklabels(movies, rotation=45, ha='right')  # Rotate x-axis labels for better readability
ax.legend()

# Showing the appearance of the  plot
plt.tight_layout()  # Adjust layout to fit labels
plt.show()
````

## Data visualization for df2

````python
#Split genres and count occurrences
genre_series = df1['genres'].dropna().str.split(',', expand=True).stack()
genre_counts = genre_series.value_counts()

# Plotting
fig, ax = plt.subplots(figsize=(14, 8))

# Plot the number of titles by genre as a bar graph
bars = ax.bar(genre_counts.index, genre_counts.values, color='blue');

# Add labels and title
ax.set_xlabel('Genre')
ax.set_ylabel('frequency of genres')
ax.set_title('Genre Frequency Distribution')
ax.set_xticks(range(len(genre_counts.index)))
ax.set_xticklabels(genre_counts.index, rotation=45, ha='right')


# Show plot
plt.tight_layout()  # Adjust layout to fit labels
plt.show()
````
